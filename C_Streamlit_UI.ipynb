# C. Streamlit UI – Code Logic / Flow

```
import streamlit as st
import joblib
import numpy as np

model = joblib.load("placement_model.pkl")

st.title("AI Career Counseling & Placement Prediction")

cgpa = st.number_input("CGPA", 0.0, 10.0, 7.0)
internship = st.selectbox("Internship", ["Yes", "No"])
aptitude = st.slider("Aptitude Score", 0, 100)
skills = st.slider("Skills Score", 0, 100)

internship_val = 1 if internship == "Yes" else 0

if st.button("Predict"):
    input_data = np.array([[cgpa, internship_val, aptitude]])
    pred = model.predict(input_data)[0]
    prob = model.predict_proba(input_data)[0][1]

    result = "Placed" if pred == 1 else "Not Placed"

    st.success(f"Result: {result}")
    st.write(f"Probability: {round(prob*100,2)}%")
```
